```python
import json
```

```python
# use of 'dump'

data = {
    "name" : "Dhruv",
    "age" : 25,
    "language" : "Panda"
}

with open("new_json.json" , "w") as f:
    json.dump(data , f)
```

```python
# use of 'dumps'

json_string = json.dumps(data)
print(json_string)
```

```python
type(json_string)
```

```python
# use of 'load'

with open("new_json.json", "r") as t:
    data  = json.load(t)

print(data)
type(data)
```

```python
# use of 'loads'

json_string = '{"name":"Dhruv","language":"Python"}'
data = json.loads(json_string)

print(data)
type(data)
```
