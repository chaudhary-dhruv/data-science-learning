```python
with open("finaldata.txt",encoding='utf-8') as f:
    data = f.read()
```

```python
chunks = data.split('\n\n')
chunks = [c for c in chunks if len(c) > 3]
```

```python
chunks[4]
```

```python
def parse_chunk(chunk):
   
    chunk = chunk.strip()
    separated_chunk = chunk.split('\n')

    #username
    username = separated_chunk[0]

    #post
    no_of_posts = int(separated_chunk[1].split(" post")[0].replace(",",""))

    #follower
    followers = float(separated_chunk[2].split(" followers")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in separated_chunk[2]):
        followers = int(followers * 1000)
    elif("M" in separated_chunk[2]):
        followers = int(followers * 1000000)
    else:
        followers = int(followers)

     #following   
    following = int(separated_chunk[3].split(" following")[0].replace(",",""))

    #name
    name = separated_chunk[4]

    #type_of_page & bio
    if(len(separated_chunk) > 5):
        type_of_page = separated_chunk[5]
        bio = '\n'.join(separated_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""

    #print(username , no_of_posts , followers , following , name , type_of_page , bio , sep='\n')  
    return{"username":username ,
           "no_of_post":no_of_posts , 
           "followers":followers ,
           "following":following , 
           "name":name , 
           "type_of_page":type_of_page , 
           "bio":bio}
```

```python
all_chunk = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunk.append(parsed_chunk)
```

```python
import json
```

```python
s = json.dumps(all_chunk , indent=4)

with open("final_data.json" , "w") as f:
    f.write(s)
```
