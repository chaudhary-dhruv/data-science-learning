```python
import pandas as pd
```

```python
df = pd.read_csv("bollywood_data.csv")
```

```python
df
```

## Column selection

```python
df["Actor"]
```

```python
df[['Actor' , 'IMDb']]
```

## Row selection

```python
df.loc[0]                # First row (by label)
```

```python
df.iloc[0]               # First row (by position)
```

```python
df.loc[0, "Actor"]        # Value at row 0, column 'Name'
```

```python
df.iloc[0, 1]            # Value at row 0, column at index 1
```

### slicing of row and column

```python
df.loc[0:2, ["Actor", "IMDb"]]   # Rows 0 to 2, selected columns
```

```python
df.iloc[0:2, 0:2]              # Rows and cols by index position
```

# Fast Access : .at and .iat

```python
df.at[0 , "Actor"]
```

```python
df.iat[0,0]
```

```python
df
```

# Filtering with Conditions

```python
df[df['IMDb'] > 7]
```

```python
#  Multiple Conditions (AND / OR)

df[(df["IMDb"] > 5) & (df["Genre"] == "Action")]
```

```python
df[(df["Actor"] == "Rajkummar Rao") | (df["Year"] > 2020)]
```

# Querying with .query()

```python
df.query("IMDb > 5 and Year > 2020")
```

```python
df.query("Actor == 'Aamir Khan'")
```

```python
df.query("`BoxOffice(INR crore)` > 500")
```

```python
df.query("`BoxOffice(INR crore)` > 500")
```

```python
good_IMDb_rating = 8

df.query("IMDb > @good_IMDb_rating")
```

```python
df.query("Year > 2018 & IMDb > 7")
```

```python
df.query("2020 < Year < 2024")
```

```python
filtered_data = df.query("2020 < Year < 2024")
```
