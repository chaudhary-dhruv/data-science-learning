```python
df2 = df.sort_values(["Year" , "IMDb"] , ascending=False).copy()

# year wise descending first , if year is same then descending the IMDb 
```

```python
df2.reset_index(drop = True , inplace=True)                    # drop function use to drop the "Index" column
                                                               # inplace: save the change in original dataframe 
```

```python
df2["Rank"] = df2["IMDb"].rank(ascending= False , method="dense")

# method = "dense" : give the same rank if condition is equal for some value
```
