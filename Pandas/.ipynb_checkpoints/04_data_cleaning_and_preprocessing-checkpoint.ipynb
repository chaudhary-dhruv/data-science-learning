```python
import pandas as pd
```

```python
df = pd.read_csv("data_cleaning_sample.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# return the all raw , that have not null value

df.dropna()
```

```python
# return the all column , that have not null value

df.dropna(axis = 1)
```

# Fill missing value

```python
# Replace missing value with 0

df.fillna(0)
```

```python
# Replace missing value with mean of that column

df['Age'].fillna(df["Age"].mean())
```

```python
# Foreard Fill

df.ffill()
```

```python
# Backward Fill

df.bfill()
```

# Detecting and Removing Duplicates

```python
df.duplicated()                        # True for duplicates Row
```

```python
df.drop_duplicates()                  # Remove duplicate rows
```

```python
#Check based on specific columns:

df.duplicated(subset=["Name", "Age"])
```

# String Operations with .str

```python
df['Name'].str.lower()                  # Converts all names to lowercase.
```

```python
df['City'].str.contains("delhi" , case = False)
```

```python
# Outputs a pandas Series where each element is a list of strings (the split parts).
#This is where a Python list comes into play, but the outer object is still a pandas Series.

df["Email"].str.split("@")
```

# Type Conversions with .astype()

```python
df2 = df.dropna().copy()
```

```python
df2['Age'] = df2['Age'].astype('int32')
```

```python
df2.info()
```

```python
df2
```

# Applying Functions

## .apply() → Apply any function to rows or columns

```python
# Both are same

df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x >= 23 else "Minor")
```

```python
def isMinor(x):
    return "Adult" if x >= 23 else "Minor"
df2["Age Group"] = df2["Age"].apply(isMinor)
```

```python
df2
```

## .map() → Element-wise mapping for Series

```python
gender_map = {"M": "Male", "F": "Female"}       # Every element that present in dataset , then should be in dictionary
df2["Gender"] = df2["Gender"].map(gender_map)
```

```python
df2
```

## .replace() → Replace specific values

```python
df["City"].replace({"Delhi": "New Delhi", "Mumbai": "New Mumbai"})
```
