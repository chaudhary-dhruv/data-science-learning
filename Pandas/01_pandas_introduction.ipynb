# DataFrame - 2D labeled Table

```python
data = {
    "Name" : ["Dhruv" , "Jay" , "Ganesh"],
    "Age" : [25 , 20 , 44],
    "City" : ["Ahmedabad" , "Pune" , "Mumbai"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```
