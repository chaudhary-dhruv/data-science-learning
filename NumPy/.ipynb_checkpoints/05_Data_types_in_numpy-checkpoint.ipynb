### downcasting to save the memory

#### Ex. In small data like age value , where 'int64' is not best option for memory so we need to downcast the array in 'int32'

```python
original_array = np.array([1,2,34,5,9,6])
```

```python
original_array.dtype
```

```python
original_array.astype('int32')
```

# check the memory usage

```python
arr_int64 = np.array([1,2,33,4,56,77] , dtype='int64')
arr_int32 = np.array([1,2,33,4,56,77] , dtype='int32')
```

```python
arr_int64.nbytes
```

```python
arr_int32.nbytes
```
